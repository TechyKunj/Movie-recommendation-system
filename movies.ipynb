```python
df = df[['title','overview','genres','tagline','vote_average','popularity']]
```

```python
df = df.dropna(subset=['title'])
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['tagline']
```

```python
#Remove punctuation and stop words
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Lowercase
    text = str(text).lower()
    # Remove punctuation and non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize and remove stop words, then lemmatize
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
```

```python
indices
```

```python
# Now we will convert into vectors
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,stop_words='english',ngram_range=(1,2))
```

```python
tfidf_metrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_metrix
```

```python
# Cosine Similarity
# graph based algo but here we will use cosine similarity
# and we have a formula for cosine similarity
# cos(theta) = A.B/||A|| ||B||
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
    if title not in indices:
        return "Movie not found."
    idx = indices[title]

    sim_score = cosine_similarity(tfidf_metrix[idx], tfidf_metrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
```

```python
recommend('Jumanji')
```

```python
import pickle
pickle.dump(tfidf_metrix,open('tfidf_metrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
df.to_pickle('df.pkl')
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
